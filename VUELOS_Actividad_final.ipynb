Concatenamos ambas listas mediante la funciÃ³n .concat() mediante axis=0 por que vamos a unir las tablas una debajo de la otra. No usamos merge por que no tienen datos en relaciÃ³n. Ya que cada fila representa una persona que pago in ticket para viajar de una ciudad a otra.

Este nuevo dataframe es asignada a la variable df3

```python
df3 = pd.concat([df1,df2],axis=0)
df3
```

Convertimos los datos de la columna a tipo entero (int), pero primero debemos reemplazar la ',' y quitar espacios para poder lograrlo.

```python
df3['price']=df3['price'].str.replace(',','').str.replace(' ','')
df3['price']=df3['price'].astype(int)

df3
```

# Limpieza de columna stop

```python
df3['stop'].value_counts() #Observamos que tienen multiplos caracteres que no permiten el entendimiento de los datos.
```

Mediante la funciÃ³n .replace() realizamos la limpieza de la columna "stop"

```python
df3['stop']=df3['stop'].str.replace('\n','')
df3['stop']=df3['stop'].str.replace('\t','')
df3['stop']=df3['stop'].str.replace('1-stopVia','')
df3['stop']=df3['stop'].str.replace('1-stop','Una Escala')
df3['stop']=df3['stop'].str.replace('non-stop','Directo')
df3['stop']=df3['stop'].str.replace('2+-stop','Dos Escalas')

df3
```

# Limpieza columna date y volverla tipo datetime.

```python
df3['date'].unique()
```

```python
df3.info() # Como podemos observar es un tipo str la columna date
```

```python
df3['date']=pd.to_datetime(df3['date'], format = '%d-%m-%Y') # Convertimos la columna 'date' en variable datetime.
df3.info()
```

```python
#Separamos el datetime esto con el proposito de darle mayor analisis mediante las graficas.
```

```python
df3['Year']=df3['date'].apply(lambda x:x.year)
df3['Month']=df3['date'].apply(lambda x:x.month)
df3['Day']=df3['date'].apply(lambda x:x.day)

df3
```
